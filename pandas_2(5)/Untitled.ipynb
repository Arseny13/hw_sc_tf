3

Среди стран, которые за всю историю Олимпийских игр завоевали более 50 комплектов золотых наград, назовите страну, у которой доля спортсменов, завоевавших золотые медали, по отношению ко всем отправленным на Олимпийские игры спортсменам является максимальной? Учитывайте медали одного и того же спортсмена, а также участие одного и того же спортсмена на разных Олимпийских играх как разные события

Ответ запишите латиницей

```python
list_county = df.groupby('NOC').agg({'gold': 'sum'}).sort_values('gold', ascending=False)[:37].index.to_list()
```

```python
def more_fifty(value):
    if value in list_county:
        return 1
    return 0
```

```python
df['qwer'] = df['NOC'].map(more_fifty)
```

```python
df1 = df[df['qwer'] == 1]
```

```python
df1.groupby('NOC').agg({'gold':lambda x: x.sum() / x.count()}).sort_values('gold', ascending=False)[:1]
```

Назовите имя и фамилию самого титулованного спортсмена Олимпийских игр (из представленного датасета)?

5


Какая максимальная разница в возрасте была у победителей (в разных видах спорта) одной сборной в рамках одних Олимпийских игр?

В ответе введите только число

```python
df[df['medal_int']==1].groupby('NOC').agg({'Age': lambda x: x.max() - x.min()}).sort_values('Age', ascending=False)[:1]
```

```python
df[df['gold']==1].groupby(['Team', 'Games']).agg({'Age': lambda x: x.max() - x.min()}).sort_values('Age', ascending=False)[:3]
```
